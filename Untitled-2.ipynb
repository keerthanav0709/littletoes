```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import cohere
import webbrowser
from threading import Timer

# Initialize Cohere client with API key
co = cohere.Client('KqeD0HJWV19p8X7AeXNNej2sY3GqEcwyP01wyrR4')  # Replace with your actual API key

# Function to get Cohere chatbot response
def get_cohere_response(user_input, chat_history):
    prompt = ""
    
    # Add chat history to the prompt
    for chat in chat_history:
        prompt += f"User: {chat['user']}\nBot: {chat['bot']}\n"

    prompt += f"User: {user_input}\nBot:"

    # Get the Cohere response
    response = co.generate(
        model='command-xlarge',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7
    )

    return response.generations[0].text.strip()

# Function to check if the question is related to education
def is_education_related(user_input):
    education_keywords = [
    # General Education
    "education", "school", "university", "college", "learning", "study", 
    "course", "subject", "teacher", "student", "exam", "homework", 
    "degree", "academic", "class", "lesson", "tutorial", 

    # Knowledge & Skills Acquisition
    "knowledge", "training", "skills", "development", "practice", 
    "computation", "understanding", "problem solving", "innovation", 
    "research", "creativity", "critical thinking", "intellectual", 

    # STEM (Science, Technology, Engineering, Mathematics)
    "science", "technology", "engineering", "mathematics", "physics", 
    "chemistry", "biology", "data science", "computer science", 
    "artificial intelligence", "machine learning", "coding", "programming", 
    "robotics", "calculus", "statistics", "quantum", "biotech", 

    # Humanities & Social Sciences
    "history", "philosophy", "psychology", "sociology", "literature", 
    "art", "music", "culture", "politics", "geography", "economics", 
    "anthropology", "ethics", "linguistics", "theology", "education policy", 

    # Business & Vocational Skills
    "business", "economics", "finance", "marketing", "entrepreneurship", 
    "management", "project management", "leadership", "communication", 
    "public speaking", "negotiation", "workshop", "seminar", 

    # Research & Methodology
    "research", "thesis", "dissertation", "case study", "fieldwork", 
    "analysis", "survey", "methodology", "data analysis", "hypothesis", 

    # Technology & Innovation
    "technology", "software", "innovation", "computing", "information systems", 
    "networking", "cybersecurity", "web development", "cloud computing", 

    # Life Skills & General Knowledge
    "life skills", "self-improvement", "personal development", 
    "productivity", "goal setting", "learning strategies", "motivation", 
    "time management", "problem solving", "critical thinking", 

    # Specific Knowledge Areas
    "medicine", "law", "architecture", "design", "ecology", "environmental science", 
    "geology", "physics", "astronomy", "chemistry", "machine learning", 
    "coding", "robotics", "statistics", "ethics", "mathematics", "philosophy"
]

    
    # Check if any education keyword is in the user's input
    return any(keyword in user_input.lower() for keyword in education_keywords)

# Initialize Dash app
app = dash.Dash(__name__)

# App layout with white and navy blue theme
app.layout = html.Div(style={'backgroundColor': '#FFFFFF', 'color': '#003366', 'padding': '10px'}, children=[
    html.H1("Cohere Chatbot", style={'textAlign': 'center', 'color': '#003366'}),
    
    # Chat history display
    dcc.Textarea(
        id='chat-history',
        value='Welcome to the chatbot! Ask me anything related to education.',
        style={'width': '100%', 'height': 300, 'backgroundColor': '#F0F0F0', 'color': '#003366', 'border': '1px solid #003366', 'padding': '10px'},
        readOnly=True
    ),
    
    # User input
    dcc.Input(
        id='user-input',
        type='text',
        placeholder='Type your message...',
        style={'width': '90%', 'padding': '10px', 'backgroundColor': '#003366', 'color': 'white', 'border': '1px solid #003366'}
    ),
    
    # Send button
    html.Button('Send', id='send-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#003366', 'color': 'white'}),
    
    # Clear button
    html.Button('Clear', id='clear-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#003366', 'color': 'white'}),
    
    # Hidden Div for storing chat history
    dcc.Store(id='chat-store', data=[])
])

# Callback to update chat and interact with Cohere
@app.callback(
    [Output('chat-history', 'value'), Output('chat-store', 'data'), Output('user-input', 'value')],
    [Input('send-btn', 'n_clicks'), Input('clear-btn', 'n_clicks')],
    [State('user-input', 'value'), State('chat-store', 'data')]
)
def update_chat(n_clicks_send, n_clicks_clear, user_input, chat_history):
    if n_clicks_clear > 0:
        # Clear the chat
        return 'Welcome to the chatbot! Ask me anything related to education.', [], ''
    
    if n_clicks_send > 0 and user_input:
        # Check if the question is education-related
        if is_education_related(user_input):
            # Get the bot response if the question is education-related
            bot_response = get_cohere_response(user_input, chat_history)
        else:
            # Respond with a message if the question is not education-related
            bot_response = "I can only help with education-related questions. Please ask something related to education."

        # Update chat history with user input and bot response
        chat_history.append({"user": user_input, "bot": bot_response})
        
        # Format chat for display
        chat_display = ""
        for chat in chat_history:
            chat_display += f"User: {chat['user']}\nBot: {chat['bot']}\n\n"
        
        # Clear user input field after sending
        return chat_display, chat_history, ''
    
    return dash.no_update, dash.no_update, dash.no_update

# Function to open the browser with the correct port
def open_browser(port):
    webbrowser.open_new(f'http://127.0.0.1:{port}/')

# Run the app
if __name__ == '__main__':
    port = 8051  # Set the port number
    Timer(1, open_browser, [port]).start()  # Pass the port to the open_browser function
    app.run_server(debug=False, use_reloader=False, port=port)
```
